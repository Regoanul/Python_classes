### List Methods

```python
nums = [3, 45, 3, 7, 78, 788,6]

nums.append(99) # add number at the last
nums.insert(1, -99) # add number at index
nums.extend([3,4,5]) # this list added in the last of nums list
nums.count(3)
nums.index(7) # 7's index number
nums.pop() # remove last value, if we provide index value then they remove this value.
nums.remove(45) # remove the number
nums.reverse()
len(nums) # provide list's length, also use in string
nums.sort() # order, small to big
nums.sort(reverse=True) # from big number to small number 
nums.clear() # list without value
```

### Shallow copy vs Deep copy

```python
nums = [2, 3, 4, 5, 6, 7]
num3 = nums # shallow copy
num3.append(99) # 99 added in both num3 and nums
nums
```

```python
num4 = nums.copy()
num4.append(33) # Deep copy
nums
```

### Set

```python
s1 = {6, 2, 3, 5, 8, 9, 10, 9}
s1.add(33) # add 33  at the last
s1.discard(9) # remove 9, if 9 is not in set no error will given
s1.remove(2) # remove 2, if 2 is not in set error will given
s1.pop() # remove last number
s1
```

```python
s1 = {2, 3, 4, 5, 6}
s2 = {7, 3, 4, 8, 9}
s1.union(s2)
s1.intersection(s2)
s1.difference(s2) 
s2.difference(s1)
s1.symmetric_difference(s2) # no common value
s1.symmetric_difference_update(s2) # symmetric_difference value s1 a update korbe.
s1
```

### Dictionary

```python
students = {1: "John", 2: "Alexa", 3: "Tessa"}
students[2] = "Alexa Smith" # update 2nd name
students[4] = "Hardin" # add name 
students
```

```python
students.get(4) # 4th name shown
students.keys() # return all keys
students.values() # return all values
students.items() # return keys and values in pair
# students.pop(2) # if run 2nd time give error
# students.popitem() # last pair pop korbe

students.update({5: "A", 6: "B"})
students.fromkeys([7,8,9], "NA") 
students.setdefault(10, "C") # add korbe like 4
students
```
